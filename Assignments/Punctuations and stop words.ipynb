```python
def remove_punctuation(text):
    no_punctuation = []
    
    for word in text:
        w = re.sub(r'[^\w\s]','',word) #remove everything except words and spaces
                                        
        w = re.sub(r'\_','',w)
        no_punctuation.append(w)
        sentences_no_punctuation = (" ").join(no_punctuation)
    
    return sentences_no_punctuation
```

```python
from gensim.parsing.preprocessing import remove_stopwords #import gensim
```

```python
#Function from book to convert docx to string
def docx_to_string(filename):
    doc = docx.Document(filename)
    fullText = []
    for para in doc.paragraphs:
        fullText.append(para.text)
    return ' '.join(fullText)
```

```python
start = time.time()
os.scandir(path=p)
cleaned_text = []

for file in glob.glob('*.*'):
    if file.endswith('.txt'):
        f = open(file, "r",encoding="utf-8")
        print(".txt")
        text_string = f.read()

    elif file.endswith('.docx'):
        print(".docx")
        text_string =  docx_to_string(file)
        
    
    else: #If not .txt or .docx
        print("Unknown File Type")

    #Parse the text of punctuation and stop words
    remove_punctuation_string = remove_punctuation(text_string.split()) #Remove Punctuation
    sentence_nostop = remove_stopwords(remove_punctuation_string) #Remove Stop Words
    cleaned_text.append([file,sentence_nostop.lower()])
```

```python
#Summary in text_summaries
for summary in cleaned_text:
    new_string = []

    for string in summary[1].split():
        if string.startswith("www"):
            new_string.append("")
        elif (string.isdigit()) and (len(string)>7):
            new_string.append("")
        else:
            new_string.append(string)

    cleaned_text[1] = (" ").join(new_string)
```
