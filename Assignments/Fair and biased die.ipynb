```python
import random as rd
import pandas as pd
from numpy.random import choice
import seaborn
import matplotlib.pyplot as plt

number=int(input('Enter number of simulations')) # to enter number of simulations value from user
fair=rd.choices([1,2,3,4,5,6],k=number)
biased=rd.choices([1,2,3,4,5,6],weights=[.15,.15,.15,.15,.15,.25],k=number)
print('Fair Die',fair,'Biased Die',biased)
```

```python
df=pd.DataFrame({'Fair Die' : fair,'Biased Die' : biased})
df.plot.bar()
plt.ylabel('Number on Dice',fontsize=11)
plt.xlabel('Turn',fontsize=11)
plt.title('Turn Vs Number on Dice')
```

```python
p=plt.hist(fair), 
p=plt.title('Fair Die')
p=plt.ylabel("Count")
p=plt.xlabel("Number on Dice")
p=plt.show()

q=plt.hist(biased), 
q=plt.title('Biased Die')
q=plt.ylabel("Count")
q=plt.xlabel("Number on Dice")
q=plt.show()
```
