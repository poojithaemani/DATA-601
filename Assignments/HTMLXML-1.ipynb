```python
import html 
from bs4 import BeautifulSoup  
import pandas as pd
```

```python
%%time
# Open the XML file containing HTML content for reading
with open("xml_containing_html.xml", 'r') as xmlfile:
    # Read the XML content into xmldoc
    xmldoc = xmlfile.read()

# Unescape HTML entities in the XML content
data = html.unescape(xmldoc)

# Open a new file for writing the unescaped HTML data
with open('output.html', 'w', encoding='utf-8') as output:
    # Write the unescaped HTML data to the file
    output.write(data)

# Open the newly created HTML file for reading
with open('output.html') as htmlfile:
    # Parse the HTML content using BeautifulSoup with 'html.parser'
    parsedData = BeautifulSoup(htmlfile.read(), 'html.parser')

# Initialize a counter for HTML pages
htmlCount = 0

# Loop through all <html> elements in the parsed HTML content
for href in parsedData.find_all('html'):
    # Increment the HTML page count
    htmlCount += 1

    # Initialize a counter for links within the current HTML page
    linkCount = 0

    # Loop through all <a> elements within the current <html> element
    for href in href.find_all('a'):
        # Increment the link count
        linkCount += 1

    # Print the HTML page number and the total number of links
    print("HTML Page", htmlCount, ":", linkCount, "Links")
```

```python
# Assuming you have parsedData containing the HTML content
html_tags = parsedData.find_all('html')

# Extract text content from the HTML tags
html_texts = [tag.get_text() for tag in html_tags]

# Create a pandas DataFrame with a single column named 'HTML'
df = pd.DataFrame({'HTML': html_texts})

# Print the DataFrame
print(df)

```
